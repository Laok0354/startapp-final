```python
indices = pd.Series(df_final.index, index=df_final['name']).drop_duplicates()
```

```python
def get_recommendations(name, cosine_sim=cosine_sim):
    idx = indices[name]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:20]

    project_indices = [i[0] for i in sim_scores]

    recommended_projects = [{"projectId": int(df_final['id'].iloc[i])} for i in project_indices]
    """recommended_project_ids = df_final['id'].iloc[project_indices].tolist()
    result_projects = [recommended_project_ids] """
    
    return recommended_projects
```

```python
import json
import time

# Retry function with exponential backoff
def retry_request(url, data, headers, retries=3, delay=1):
    for i in range(retries):
        try:
            response = requests.post(url, data=json.dumps(data), headers=headers)
            return response
        except requests.exceptions.ConnectionError:
            print(f"Connection error. Retrying in {delay} seconds...")
            time.sleep(delay)
            delay *= 2
    return None

url = "http://localhost:3000/recommendations/recommend"

res = get_recommendations("cat")

datos_json = {
    "id": 56,
    "projectsWithAffinities": res
} 

encabezado = {
    "Authorization": "Bearer",
    "Content-Type": "application/json",
    "x-api-key": "REYDEREDES"
}

print("Hola")
respuesta = retry_request(url, datos_json, encabezado)

print("Hola")
if respuesta is not None and respuesta.status_code == 200:
    print('Solicitud exitosa. Respuesta del servidor:')
    print(respuesta.json())
else:
    print(f'Error en la solicitud. Cdigo de estado: {respuesta.status_code}')
    print(respuesta.text if respuesta else "No response")

""" print(datos_json)

encabezado = {
    "Authorization": "Bearer",
    "Content-Type": "application/json",
    "x-api-key": "REYDEREDES"
}

print("Hola")
respuesta = requests.post(url, data=json.dumps(datos_json), headers=encabezado)


print("Hola")
if respuesta.status_code == 200:
    print('Solicitud exitosa. Respuesta del servidor:')
    print(respuesta.json())
else:
    print(f'Error en la solicitud. Cdigo de estado: {respuesta.status_code}')
    print(respuesta.text) """
```
